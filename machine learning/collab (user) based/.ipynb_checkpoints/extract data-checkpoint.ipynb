```python
qty = int(9906*0.8)
qty
```

```python
initializer = tf.keras.initializers.RandomNormal(mean=0., stddev=1.)
YU = tf.keras.Sequential([
            tf.keras.layers.Dense(units=10, input_shape=[1], activation="relu", kernel_initializer=initializer, kernel_regularizer=regularizers.l2(0.01)),
#             tf.keras.layers.Dense(units=32,activation="relu"),
    
        ])
YU.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
YU.save_weights('models.h5')
```

data process

```python
xs = np.zeros([9906])
ys = matris

for i in range(9906):
    xs[i] = i #initial index
```

```python
matris.shape[0]
```

### split data, train & test

```python
ys_train = np.zeros([136, qty])
ys_test = np.zeros([136, 9906-qty])
```

```python
xs_train, xs_test = xs[:qty], xs[qty:]

for i in range(matris.shape[0]):
    ys_train[i], ys_test[i] = ys[i][:qty], ys[i][qty:]
```

```python
print(xs_train.shape)
print(ys_train.shape)
```

```python
test_loss = np.zeros([136])
weights = np.zeros([136, 10])
biases = np.zeros([136, 10])
```

```python
for i in range(5):

    clear_output(wait=True)
    
    print("iter ", i)

    YU.fit(xs_train, ys_train[i], epochs=2)
    test_loss[i] = YU.evaluate(xs_test, ys_test[i])

    weights[i] = YU.layers[0].get_weights()[0]
    
    for j in range(10):
        biases[i][j] = YU.layers[0].get_weights()[1][j]
        
    YU.load_weights('models.h5')
```

```python
matris_int = matris.astype(int)
baris = baris.astype(int)
baris[1]
```

```python
# Loading the PCA algorithm
pca_2 = PCA(n_components=1)
pca_2
```

```python
pca_2.fit(baris)
```

```python
pca_2.explained_variance_ratio_
```

```python
baris_trans_2 = pca_2.transform(baris)
baris_trans_2
```

```python
baris_reduced_2 = pca_2.inverse_transform(baris_trans_2)
baris_reduced_2
```
