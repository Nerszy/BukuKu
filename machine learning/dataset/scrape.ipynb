```python
import io
df = pd.read_csv(io.BytesIO(uploaded['books.csv']))# Dataset is now stored in a Pandas Dataframe

# df = pd.read_csv("C:\\SHAFA\\COOLYEAH\\Semester 6\\Bangkit\\Capstone\\dataset\\books.csv") #sesuaikan path
# dataframe to list
value = df.values.tolist()
```

```python
list_of_lists = []
```

```python
first_index = 671
until_index = 1000
for item in (value[first_index:until_index]):
    id = str(item[1])
    # open goodreads
    playbook_url = "https://www.goodreads.com/book_link/follow/8036?book_id=" + id + "&search=title&source=compareprices"
    source = urlopen(str(playbook_url))
    soup = BeautifulSoup(source, 'html.parser')
    playbook = soup.find(class_="Si6A0c")
    # open playbook
    try:
        url_playbook = "https://play.google.com" + playbook['href']
    except:
        error.append(item[1])
        continue

    # OPEN URL PLAYBOOK
    source = urlopen(url_playbook)
    soup = BeautifulSoup(source, 'html.parser')

#     url_image
    img = soup.find_all("img")[1]
    image = img['srcset']
    url_image = image[:len(image)-3]

#     price
    temp_price = soup.find(itemprop="price")
    try:
      price = str(temp_price['content'])
    except:
      continue

#     synopsis
    temp_sinopsis = soup.find_all( class_="LKWyGd")
    isi = str(temp_sinopsis)
    synopsis = isi[22:-8]
    synopsis = synopsis.replace(";", ",")
    
#     title
    title = soup.find("h1")
    print(title.text) #progress output
    title = title.text

    list_of_lists.append([item[1], title, url_playbook, url_image, synopsis, price])
```
